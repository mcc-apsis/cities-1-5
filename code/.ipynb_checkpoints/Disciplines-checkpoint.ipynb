```python
tvalues = []
for t in topics:
    tv = {"t_id": t.pk, "title":t.title, "score": t.score}
    for wc in wcs.iterator():
        dts = DocTopic.objects.filter(
            run_id=run_id,topic=t,doc__wc=wc
        ).aggregate(
            total = Sum('score')
        )
       

        twcv = {
            "wc_id" : wc.id,
            "oecd" : wc.oecd,
            "oecd_fos_text" : wc.oecd_fos_text,
            "wc_score" : dts['total']
        }
        twcv.update(tv)
        #twcv['text'] = wc.text
        #twcv['oecd'] = wc.oecd
        #twcv['oecd_fos_text'] = wc.oecd_fos_text
        #twcv['score'] = dts['total']
        #print(wc.__dict__)
        #print(twcv)    
        #print(t.title)
        #print(wc.text)
        
        tvalues.append(twcv)
        
```
